```python
docs_info={}
with open(r"pickles/docs_info.pickle", "rb") as f:
    docs_info = pickle.load(f, encoding="latin1")
```

```python
keywords={}
empty=0
tot=0
with open('../search_engine/results/academic/test/academic_results_4_features_ndcg50_.02.txt') as f:
    for idx, line in enumerate(f):
        l=line.split("\t")
        #if l[0]=='9' and idx%100<10:
        if idx%100<10:
            print(l[0], l[1], l[2])
            kws=docs_info[l[1]]['keyPhrases']
            if kws==' ':
                empty+=1
            tot+=1
            for kw in kws:
                if kw in keywords:
                    keywords[kw]+=1
                else:
                    keywords[kw]=1
```
